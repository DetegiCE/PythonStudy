```python
import numpy as np

#normal : 정규분포에서 표본 추출
print(np.random.normal(size=5))
print(np.random.normal(size=5))

#동일한 값을 추출하고자 하는 경우에 seed를 이용해 난수 초기값 부여
np.random.seed(seed=100)
print(np.random.normal(size=5))
print(np.random.normal(size=5))
np.random.seed(seed=100)
print(np.random.normal(size=5))
```

```python
print(np.random.normal(size=(2, 3)))
print(np.random.normal(size=(2, 3, 4)))
```

```python
#binomial : 이항분포
print(np.random.binomial(n=1, p=0.5, size=20))

#hypergeometric : 초기하분포
print(np.random.hypergeometric(ngood=5, nbad=20, nsample=5, size=100))

#poisson : 포아송분포
print(np.random.poisson(lam=20, size=100))
```

```python
mu, sigma = 0.0, 3.0
print(np.random.normal(mu, sigma, size=100))
```

```python
print(np.random.normal(loc=0, scale=3, size=100)) #loc = mean, scale = variance
print(np.random.randn(100))
```

```python
#standard_t : t분포
print(np.random.standard_t(df=3, size=100))

#uniform : 균등분포
print(np.random.uniform(low=0.0, high=10.0, size=100))

#randint : 이산형 균등분포에서 정수형 표본 추출
print(np.random.randint(low=0, high=10, size=100))
print(np.random.randint(low=0, high=11, size=100))

#f : f분포
print(np.random.f(dfnum=5, dfden=10, size=100))

#chisquare : 카이제곱분포
print(np.random.chisquare(df=2, size=100))

#gamma : 감마분포
print(np.random.gamma(shape=2, scale=2, size=100))
```

```python
#permutation, shuffle : 임의로 순서를 바꿔서 추출
a = np.array([1, 4, 7, 2, 6, 9])
print(np.random.permutation(a))
print(a)

a = np.array([1, 4, 7, 2, 6, 9])
print(np.random.shuffle(a))
print(a)
```
